```python
import re
```

```python
with open('/workspaces/ivy/ivy/array/elementwise.py', 'r') as f:
    code = ''.join(f.readlines())
```

```python
pattern = re.compile(r'"""[\w\W]*?"""')
new_line = re.compile(r'\n')
```

```python
for m in pattern.finditer(code):
    # print(m.start(0))
    start_line = len(new_line.findall(code, 0, m.start(0)))+1
    end_line = len(new_line.findall(code, 0, m.end(0)))+1
    print ('content={}, start={}, end={}'.format(m.group(0), start_line, end_line))
```
